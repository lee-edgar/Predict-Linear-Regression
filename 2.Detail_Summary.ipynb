### 채널별, 소재별, 시간대 컬럼으로 세분화작업

```python
df_result = pd.read_excel("./Datas/df_result_fn.xlsx")
df_result.tail()
```

----

## 1. 채널별분석

### -채널별 광고송출량 및 후원실적

```python
by_ch = df_result.groupby("Channel_x").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_ch["ROI"] = round(by_ch["후원금액"]*12 / by_ch["Net Price"], 2)
by_ch
```

### -채널별 상관계수

```python
ch_corr = pd.read_excel("./Datas/coref.xlsx", "Channel")
ch_corr
```

## 2. 시간대별분석

### -시간별 광고송출량 및 후원실적

```python
by_daypart = df_result.groupby("Daypart_x").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_daypart["ROI"] = round(by_daypart["후원금액"]*12 / by_daypart["Net Price"], 2)
by_daypart
```

### -시간대별 상관계수

```python
dp_corr = pd.read_excel("./Datas/coref.xlsx", "Daypart")
dp_corr
```

## 3. 광고소재별

### -광고소재별 광고송출량 및 후원실적

```python
by_cre = df_result.groupby("Creative_x").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_cre["ROI"] = round(by_cre["후원금액"]*12 / by_cre["Net Price"], 2)
by_cre
```

### -광고소재별 상관계수

```python
cr_corr = pd.read_excel("./Datas/coref.xlsx", "Creative")
cr_corr
```
