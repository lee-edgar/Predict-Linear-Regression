## 크리에이티브별 카테고리

```python
cr_quality = pd.read_excel("./Datas/cr_quality.xlsx")
cr_quality.tail()
```

```python
### 전체데이터와 크리에이티브별카테고리합치기
```

```python
cr_df = cr_quality[["Creative_x", "Category", "Material", "Asklevel", "Model"]]
cr_df
```

```python
cr_merged = pd.merge(df_result, cr_df, on='Creative_x', how='outer')
cr_merged.tail(5)
```

## 광고별분석

### 광고소재별

```python
by_ch = cr_merged.groupby("Creative_x").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_ch["ROI"] = round(by_ch["후원금액"]*12 / by_ch["Net Price"], 2)
by_ch.head(10)
```

### 광고소재의장르별

```python
by_cate = cr_merged.groupby("Category").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_cate["ROI"] = round(by_cate["후원금액"]*12 / by_cate["Net Price"], 2)
by_cate
```

### 광고소재의 모델유무별

```python
by_model = cr_merged.groupby("Model").sum().sort_values(by='I/B콜수', ascending=False).reset_index()
by_model["ROI"] = round(by_model["후원금액"]*12 / by_model["Net Price"], 2)
by_model
```
